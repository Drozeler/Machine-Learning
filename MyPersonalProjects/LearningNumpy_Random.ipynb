```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = random.randint(100)
print(x)
```

```python
x = random.rand() #Random float
print(x)
```

```python
x = random.randint(10, size=(5)) #A 1d array of 5 numbers in the range of 10
print(x)
```

```python
x = random.randint(10, size=(2, 5)) #A 2d array of 2 rows with 5 numbers each in the range of 10
print(x)

sorted = np.sort(x)
print(sorted)
```

```python
x = random.rand(5) #rand() function is for float, the number inside the bracket is the amount of decimal places
print(x)

y = random.rand(2,5) #2d, same thing, 2 rows 5 values each
print(y)
```

```python
x = random.randint((3,5,7,9)) #Works with arrays
print(x)
```

```python
x = random.randint((3,5,7,9), size =(2,4))
print(x)
```

```python
x = random.choice((3,5,7,9), p=[0.1,0.3,0.6,0], size=(10)) #The sum of all probability numbers should be 1.
print(x)

#Generate a 1-D array containing 10 values, where each value has to be 3, 5, 7 or 9.

#The probability for the value to be 3 is set to be 0.1

#The probability for the value to be 5 is set to be 0.3

#The probability for the value to be 7 is set to be 0.6

#The probability for the value to be 9 is set to be 0
```

```python
x = random.choice((3,5,7,9), p=[0.1,0.3,0.6,0], size=(2, 5)) #The sum of all probability numbers should be 1.
print(x)
```

```python
arr = np.array((1,2,3,4,5))
random.shuffle(arr)
print(arr)
```

```python
arr = np.array((1,2,3,4,5))
x = random.permutation(arr)
print(x) #Shuffles it but leaves the original array unchanged
print(arr)
```

```python
#Using Seaborn
sns.distplot((0,1,2,3,4,5)) #displot means distribution plot
plt.show()
```

```python
sns.distplot((0,1,2,3,4,5), hist=False) #Without histogram
plt.show()
```

```python
x = random.normal(size=(2,3))
print(x)
```

```python
x = random.normal(loc = 1, scale = 2, size=(2,3))
print(x)
#loc - (Mean) where the peak of the bell exists.

#scale - (Standard Deviation) how flat the graph distribution should be.

#size - The shape of the returned array.
```

```python
sns.distplot(random.normal(size=1000), hist=False)
plt.show()
```

```python
x = random.binomial(n = 10, p = 0.5, size = 10) #Given 10 trials for coin toss generate 10 data points:
print(x)
#n - number of trials.

#p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).

#size - The shape of the returned array.
```

```python
sns.distplot(random.binomial(n = 10, p = 0.5, size = 1000), hist = True, kde = False)
plt.show()
```

```python
sns.distplot(random.normal(loc = 50, scale = 5, size = 1000), hist = False, label = 'Normal')
sns.distplot(random.binomial(n = 100, p = 0.5, size = 1000), hist = False, label = 'Binomial')
plt.show()#Binomial vs Normal
```

```python
x = random.poisson(lam = 2, size = 10)
print(x)
#lam - rate or known number of occurrences e.g. 2 for above problem.

#size - The shape of the returned array.
```

```python
sns.distplot(random.poisson(lam = 2, size = 1000), kde = False)
plt.show()
```

```python
sns.distplot(random.normal(loc = 50, scale = 7, size = 1000),hist=False, label='normal')
sns.distplot(random.poisson(lam = 50, size = 1000),hist=False, label='poisson')
plt.show()#Normal vs Poisson
```

```python
sns.distplot(random.binomial(n = 1000, p = 0.01, size = 1000), hist = False, label = 'binomial')
sns.distplot(random.poisson(lam = 10, size = 1000), hist = False, label = 'poisson')
plt.show()
```

```python
x = random.uniform(size=(2,3))
print(x)
```

```python
sns.distplot(random.uniform(size=1000), hist = False)
plt.show()
```

```python
x = random.logistic(loc = 1, scale = 2, size = (2,3))
print(x)
#loc - mean, where the peak is. Default 0.

#scale - standard deviation, the flatness of distribution. Default 1.

#size - The shape of the returned array.
```

```python
sns.distplot(random.logistic(size=(1000)), hist = False)
plt.show()
```

```python
sns.distplot(random.normal(scale = 2, size=1000), hist = False, label = 'normal')
sns.distplot(random.logistic(size = 1000), hist = False, label = 'logistic')
plt.show()
```

```python
x = random.multinomial(n = 6, pvals = [1/6,1/6,1/6,1/6,1/6,1/6])
print(x)
#n - number of possible outcomes (e.g. 6 for dice roll).

#pvals - list of probabilties of outcomes (e.g. [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] for dice roll).

#size - The shape of the returned array.
```

```python
x = random.exponential(scale = 2, size = (2,3))
print(x)
```

```python
sns.distplot(random.exponential(size = 1000), hist = False)
plt.show()
```

```python
x = random.chisquare(df = 2, size = (2,3))
print(x)
#df - (degree of freedom).

#size - The shape of the returned array.
```

```python
sns.distplot(random.chisquare(df = 1, size = 1000),hist = False)
plt.show()
```

```python
x = random.rayleigh(scale = 2, size = (2,3))
print(x)
#scale - (standard deviation) decides how flat the distribution will be default 1.0).

#size - The shape of the returned array.
```

```python
sns.distplot(random.rayleigh(size=1000), hist = False)
plt.show()
```

```python
x = random.pareto(a = 2, size = (2,3))
print(x)
#a - shape parameter.

#size - The shape of the returned array.
```

```python
sns.distplot(random.pareto(a = 2, size = 1000),kde = False)
plt.show()
```

```python
x = random.zipf(a = 2, size = (2,3))
print(x)
#a - distribution parameter.

#size - The shape of the returned array
```

```python
sns.distplot(random.zipf(a = 2, size = 1000),kde = False)
plt.show()
```
