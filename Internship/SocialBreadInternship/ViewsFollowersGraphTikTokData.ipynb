```python
import pandas as pd

# Load the data
df = pd.read_csv('followersviews.csv')
df.head()
```

```python
import numpy as np
X = df['Followers'] # EXtract the 'Followers' column as a Series (1D)
y = df['Avg Views']
m, c = np.polyfit(X, y, 1)
```

```python
y_trend = m * X + c
y_trend_from_zero = m * X  # This starts from y = 0
y_trend_1_5 = 2.5 * y_trend  # 1.5 times the original trendline
y_trend_0_6 = 0.3 * y_trend_from_zero
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_trend, color='red', label='Trendline: y = {:.2f}X + {:.2f}'.format(m, c))
plt.plot(X, y_trend_1_5, color='green', linestyle='--', label='1.5 * Trendline')
plt.plot(X, y_trend_0_6, color='orange', linestyle='-.', label='0.6*Trendline(From y=0)')
plt.xlim(left=0)

plt.gca().xaxis.set_major_formatter(mticker.ScalarFormatter())
# ... (your existing code) ...
ax = plt.gca()
ax.ticklabel_format(style='plain', axis='x')  # Add this line here



# ... (rest of your plotting code) ...
plt.xlabel('Followers')
plt.ylabel('Avg Views')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```
